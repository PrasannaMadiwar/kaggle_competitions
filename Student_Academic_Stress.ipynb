2. **Influence of Academic and Home Pressure**
   - Increased pressure from both academic competition and family expectations leads to a significant rise in overall student stress.  


## Now lets do Some preprocessing :


```python
data['Study_environment'].fillna('Peaceful',inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['high school','undergraduate','post-graduate']])
data['Aca_stage'] = oe.fit_transform(data[['Aca_stage']])
data['Aca_stage'] = data['Aca_stage'].astype(int)
```

```python
oe1 = OrdinalEncoder(categories=[['No','Yes','prefer not to say']])
data['bad_habits'] = oe1.fit_transform(data[['bad_habits']])
data['bad_habits'] = data['bad_habits'].astype(int)
```

```python
oe2 = OrdinalEncoder(categories=[['disrupted','Peaceful','Noisy']])
data['Study_environment'] = oe2.fit_transform(data[['Study_environment']])
data['Study_environment'] = data['Study_environment'].astype(int)
```

```python
oe3 = OrdinalEncoder(categories=[['Emotional breakdown (crying a lot)','Analyze the situation and handle it with intellect','Social support (friends, family)']])
data['Coping_strategy'] = oe3.fit_transform(data[['Coping_strategy']])
data['Coping_strategy'] = data['Coping_strategy'].astype(int)
```

## Ordinal Encoding Summary

The categorical variables were transformed into numerical values using **Ordinal Encoding**.  
Each category was mapped to an integer based on a predefined order, as shown below:

- **Academic Stage (`Aca_stage`)**
  - `high school` → **0**
  - `undergraduate` → **1**
  - `post-graduate` → **2**

- **Bad Habits (`bad_habits`)**
  - `No` → **0**
  - `Yes` → **1**
  - `prefer not to say` → **2**

- **Study Environment (`Study_environment`)**
  - `disrupted` → **0**
  - `Peaceful` → **1**
  - `Noisy` → **2**

- **Coping Strategy (`Coping_strategy`)**
  - `Emotional breakdown (crying a lot)` → **0**
  - `Analyze the situation and handle it with intellect` → **1**
  - `Social support (friends, family)` → **2**

---

### Note  
This encoding preserves the **ordinal relationship** among categories where applicable (e.g., academic stage), while ensuring that all categorical data is in numeric format for further analysis and modeling.

# Now we will extract information from the Date

```python
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
```

```python
data['year'] = data['Timestamp'].dt.year
data['month'] = data['Timestamp'].dt.month
data['day'] = data['Timestamp'].dt.day
data['semester'] = np.where(data['month'].between(1, 6), 1, 2)
```

```python
data.drop('Timestamp',axis=1,inplace=True)
```

```python
data.sample(10)
```

 ### Observation on Stress Levels Across Semesters

We noticed that students tend to report **higher stress in even semesters**.  
However, this pattern may not necessarily indicate that stress is inherently higher in even semesters.  
It could also be due to a **data collection bias**, since the survey/data was taken primarily during the **even semester period**.  

➡️ To confirm this, we would need stress data from **both odd and even semesters** for comparison.
