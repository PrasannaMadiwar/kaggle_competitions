```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.sample(10)
```

```python
df.drop(['Name','PassengerId','Cabin'],axis=1,inplace=True)
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
sichar = SimpleImputer(strategy='most_frequent')
sinum = SimpleImputer(strategy='median')
ore = OrdinalEncoder()
one = OneHotEncoder()
```

```python
df.info()
```

```python

df['HomePlanet'] = sichar.fit_transform(df[['HomePlanet']]).ravel()
df['CryoSleep'] = sichar.fit_transform(df[['CryoSleep']]).astype(int).ravel()
df['Destination'] = sichar.fit_transform(df[['Destination']]).ravel()
df['VIP'] = sichar.fit_transform(df[['VIP']]).astype(int).ravel()
df['Age'] = sinum.fit_transform(df[['Age']]).ravel()
df['RoomService'] = sinum.fit_transform(df[['RoomService']]).ravel()
df['FoodCourt'] = sinum.fit_transform(df[['FoodCourt']]).ravel()
df['ShoppingMall'] = sinum.fit_transform(df[['ShoppingMall']]).ravel()
df['Spa'] = sinum.fit_transform(df[['Spa']]).ravel()
df['VRDeck'] = sinum.fit_transform(df[['VRDeck']]).astype(int).ravel()
```

```python
df.head(5)
```

```python
home =  pd.get_dummies(df['HomePlanet']).astype(int)
```

```python
dest = pd.get_dummies(df['Destination']).astype(int)
```

```python
data = pd.concat([df,home,dest],axis=1)
```

```python
data.drop(['HomePlanet','Destination'],axis=1,inplace=True)
```

```python
data.corr()
```

```python
#from ydata_profiling import ProfileReport
#profile = ProfileReport(data)
#profile.to_notebook_iframe()
```

```python
data.drop(['VIP','PSO J318.5-22'],axis=1,inplace=True)
```

```python
X = data.drop('Transported',axis=1)
y = data['Transported']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
bc = BaggingClassifier(estimator=rf)
```

```python

#bc.fit(X_train,y_train)
```

```python
#print('Voting Classifier : ',accuracy_score(y_test,bc.predict(X_test)))
```

```python
#bc.oob_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params  = {
    'n_estimators': [50, 100, 150],
    'max_samples': [0.5, 0.7, 1.0],
    'max_features': [0.5, 0.7, 1.0],
    'estimator__max_depth': [None, 5, 10],
    'estimator__min_samples_split': [2, 5, 10],
    'bootstrap': [True, False],
    'bootstrap_features': [True, False],
    'oob_score': [True, False]
}

```

```python
gs = RandomizedSearchCV(bc,param_distributions=params,cv=5,n_jobs=-1,verbose=1)
```

```python
gs.fit(X_train,y_train)
```

```python
pm = gs.best_estimator_
```

```python
print(accuracy_score(y_test,pm.predict(X_test)))
```
